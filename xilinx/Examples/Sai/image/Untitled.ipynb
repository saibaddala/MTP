```python
import socket

def receive_image(save_path, host, port):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind((host, port))
        s.listen()
        conn, addr = s.accept()
        with conn:
            with open(save_path, 'wb') as file:
                while True:
                    data = conn.recv(1024)
                    if not data:
                        break
                    file.write(data)

SAVE_PATH = '/home/xilinx/image_received.jpg'
HOST = '0.0.0.0'
PORT = 65432

receive_image(SAVE_PATH, HOST, PORT)
```
