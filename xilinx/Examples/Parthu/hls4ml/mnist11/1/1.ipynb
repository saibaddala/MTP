```python
from axi_stream_driver import NeuralNetworkOverlay
import numpy as np
import time

# Load test data
X_test = np.load('x_test.npy')
y_test = np.load('y_text.npy')  # Make sure this file is spelled correctly!

# Load the overlay (bitstream + driver setup)
nn = NeuralNetworkOverlay('design.bit', X_test.shape, y_test.shape)

# Start timer for PS-side measurement (includes DMA transfer + PL compute + result copy)
start_ps = time.time()

# Run inference on the hardware with profiling
y_hw, latency_total, throughput = nn.predict(X_test, profile=True)

end_ps = time.time()
ps_total_time = end_ps - start_ps

# Number of input samples
num_samples = X_test.shape[0]

# Per-sample latency based on reported total latency
latency_per_sample = latency_total / num_samples

# Print results
print(f"\n=== Inference Summary ===")
print(f"Samples classified        : {num_samples}")
print(f"Total PL compute time     : {latency_total:.6f} sec")
print(f"Per-sample PL latency     : {latency_per_sample:.6e} sec")
print(f"Total PS-side time        : {ps_total_time:.6f} sec")
print(f"Throughput (from overlay) : {throughput:.2f} samples/sec")

# Optional: compare how much time is spent in PS overhead
ps_overhead = ps_total_time - latency_total
ps_overhead_percent = (ps_overhead / ps_total_time) * 100
pl_percent = (latency_total / ps_total_time) * 100

print(f"\n--- Timing Breakdown ---")
print(f"PL compute time  : {pl_percent:.2f}% of total")
print(f"PS overhead      : {ps_overhead:.6f} sec ({ps_overhead_percent:.2f}% of total)")
```

```python
from axi_stream_driver import NeuralNetworkOverlay
import numpy as np
import time

# Load test data
X_test = np.load('x_test.npy')
y_test = np.load('y_text.npy')

# Load the hardware overlay
nn = NeuralNetworkOverlay('design.bit', X_test.shape, y_test.shape)

# Lists to store timing information
ps_times = []
pl_latencies = []

# Run inference one sample at a time
for i in range(len(X_test)):
    x = np.expand_dims(X_test[i], axis=0)  # Shape: (1, ...)

    # Start PS-side timer
    start_ps = time.time()

    # Run hardware inference with profiling
    y_hw, latency_pl, _ = nn.predict(x, profile=True)

    # End PS-side timer
    end_ps = time.time()

    # Calculate and store times
    ps_time = end_ps - start_ps
    ps_times.append(ps_time)
    pl_latencies.append(latency_pl)

    # Optional progress print
    if i % 1000 == 0:
        print(f"Processed {i} samples")

# Convert to numpy arrays for analysis
ps_times = np.array(ps_times)
pl_latencies = np.array(pl_latencies)

# Calculate totals and averages
avg_ps_time = np.mean(ps_times)
avg_pl_latency = np.mean(pl_latencies)
avg_ps_overhead = np.mean(ps_times - pl_latencies)
total_pl_time = np.sum(pl_latencies)

# Print summary
print(f"\n=== Per-Sample Timing Summary ===")
print(f"Total samples processed         : {len(X_test)}")
print(f"Total PL compute time           : {total_pl_time:.6f} sec")
print(f"Avg PS-side time per sample     : {avg_ps_time:.6f} sec")
print(f"Avg PL-only latency per sample  : {avg_pl_latency:.6f} sec")
print(f"Avg PS overhead per sample      : {avg_ps_overhead:.6f} sec")

print(f"\n--- Detailed Stats ---")
print(f"Min PS-side time                : {np.min(ps_times):.6f} sec")
print(f"Max PS-side time                : {np.max(ps_times):.6f} sec")
print(f"Min PL latency                  : {np.min(pl_latencies):.6f} sec")
print(f"Max PL latency                  : {np.max(pl_latencies):.6f} sec")
```
